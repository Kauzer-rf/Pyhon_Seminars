```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
gx = -5*x**2+10*x-150
```

```python
# Находим корни
```

```python
[root1, root2] = solve(gx,x)
print(f'{root1}\n{root2}')
```

```python
# Находим интервалы, на которых функция возрастает
```

```python
solve(diff(gx)>0, x,check=False)
```

```python
# Находим интервалы, на которых функция убывает
```

```python
solve(diff(gx) < 0, x, check=False)
```

```python
# Построим график
```

```python
p = plot(gx)
```

```python
# Ищем вершину 
```

```python
solve(diff(gx), x, check=False)
```

```python
# Определим промежуток, на котором f > 0
```

```python
solve(gx > 0, x,check=False)
```

```python
# Определим промежуток, на котором f < 0
```

```python
solve(gx < 0, x,check=False)
```
